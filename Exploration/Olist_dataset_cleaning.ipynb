The purpose of this project is to **clean and preprocess** Olist’s sales data, ensuring accuracy, consistency, and reliability for analysis. ## Dataset Overview
This project uses multiple datasets:
- **olist_orders_dataset.csv** – Contains order details, including purchase timestamps, customer info, and order status.
- **olist_order_items_dataset.csv** – Includes order items, pricing, and seller details.
- **olist_customers_dataset.csv** – Provides customer location data.
- **olist_products_dataset.csv** – Lists product categories and features.
- **olist_order_payments_dataset.csv** – Shows payment method distribution.
- **product_category_name_translation.csv** – Translates product categories from Portuguese to English.

These datasets together enable a comprehensive view of Olist’s sales performance, customer behavior, and regional insights.

```python
import pandas as pd

# Loading each dataset into a pandas DataFrame 
orders_df = pd.read_csv('olist_orders_dataset.csv')
items_df = pd.read_csv('olist_order_items_dataset.csv')
products_df = pd.read_csv('olist_products_dataset.csv')
customers_df = pd.read_csv('olist_customers_dataset.csv')
sellers_df = pd.read_csv('olist_sellers_dataset.csv')
payments_df = pd.read_csv('olist_order_payments_dataset.csv')
```

## Merging the Data
Since the data was split across multiple tables, merging was essential to create a unified dataset.

One method used for merging:

```python
from functools import reduce
data_frames = [orders_df,payments_df,items_df]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['order_id'],how='left'), data_frames)
merge_dftwo = pd.merge(df_merged,customers_df,on ="customer_id", how="left").merge(sellers_df,on ="seller_id", how="left").merge(products_df,on ="product_id", how="left")
merge_dftwo.head()
```

Another  way one can merge

```python
main_dataset = pd.merge(orders_df,payments_df, on ="order_id", how="left").merge(items_df,on ="order_id", how="left").merge(customers_df,on ="customer_id", how="left").merge(sellers_df,on ="seller_id", how="left").merge(products_df,on ="product_id", how="left")
main_dataset.head()
```

## Handling Missing Values
Missing values can distort analysis. They were removed with:

```python
cleaned_dataset = main_dataset.dropna()
cleaned_dataset.head()
```

## Selecting Relevant Columns for the analysis
Not all columns are useful for analysis. The dataset was trimmed to only essential features:

```python
columns_to_keep = [
    "order_id", "customer_id", "order_status", "order_purchase_timestamp",
    "order_delivered_customer_date", "payment_type", "payment_value", "payment_installments",
    "order_item_id", "product_id", "seller_id", "price", "freight_value",
    "customer_city", "customer_state", "seller_city", "seller_state",
    "product_category_name", "product_photos_qty"
]
new_selected_cols = cleaned_dataset[columns_to_keep]
print(len(columns_to_keep))
new_selected_cols.head()
```

## Translating Product Names
Since the dataset had Portuguese product names, they were converted to English:

```python
#translating product name to english using the translation csv
productname_to_english =pd.read_csv("product_category_name_translation.csv")
merged_name_sets =pd.merge(new_selected_cols,productname_to_english, on="product_category_name" ,how="left")
merged_name_sets.head()
```

```python
#converting product name to english
merged_name_sets.drop(columns=['product_category_name'], inplace = True )
merged_name_sets.rename(columns={"product_category_name_english":"product_category_name"}, inplace = True )
merged_name_sets.head()
```

## Saving the Cleaned Data
The final cleaned dataset was saved as a CSV:

```python
merged_name_sets.head(10000).to_csv('olist_dataset.csv', index=False)
```

## Converting to SQL Format
To facilitate querying, the cleaned data was converted into SQL commands:

```python
#Coverting CSV to SQL
import pandas as pd

# Read CSV file
df = pd.read_csv('olist_dataset.csv')

# Define SQL column types based on pandas dtypes
columns = df.dtypes.apply(lambda x: 'TEXT' if x == 'object' else 'NUMERIC').to_dict()

# Create table statement
create_table = f"CREATE TABLE olist_sales (\n" + \
               ',\n'.join([f"    {col} {dtype}" for col, dtype in columns.items()]) + \
               "\n);"

# Convert DataFrame to SQL INSERT statements
def escape_sql_value(value):
    """Escapes single quotes and handles NULL values."""
    if pd.isna(value):  # Handle NaN values
        return 'NULL'
    if isinstance(value, str):  # Escape single quotes in text fields
        return f"'{value.replace("'", "''")}'"
    return str(value)

values = df.map(escape_sql_value)  # Apply escape function to all values
inserts = 'INSERT INTO olist_sales VALUES\n' + \
          ',\n'.join([f"({','.join(row)})" for row in values.astype(str).values]) + \
          ';'

# Write to file with UTF-8 encoding
with open('olist_sales.sql', 'w', encoding='utf-8') as f:
    f.write(create_table + '\n\n' + inserts)

print("SQL script generated successfully!")
```
